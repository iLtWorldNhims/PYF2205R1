```python
from random import *

if randint(0,1) == 0:
    current_player  = "human"
else:
    current_player  = "computer"
        
while True:
    current_number = 0
    
    while current_number <= 21:
        print("Số hiện tại là:", current_number)
        
        if current_player  == "human":
            player_choice = int(input("Bạn chọn: "))
            while player_choice not in [1, 2, 3]:
                player_choice = int(input("Yêu cầu nhập lại! Nhập số từ 1 - 3: "))
            current_number += player_choice
            print("Số hiện tại là:", current_number)
            
            if current_number >= 21:
                print("Số hiện tại là:", current_number)
                print("Bạn đã thua cuộc! Trò chơi kết thúc.")
                break
            else:
                current_player  = "computer"
                computer_choice = randint(1,3)
                print("Máy chọn:", computer_choice)
                current_number += computer_choice

                if current_number >= 21:
                    print("Số hiện tại là:", current_number)
                    print("Bạn đã chiến thắng! Trò chơi kết thúc.")
                    break

                else:
                    current_player  = "human"
                    
        else:
            computer_choice = randint(1,3)
            print("Máy chọn:", computer_choice)
            current_number += computer_choice

            if current_number >= 21:
                print("Số hiện tại là:", current_number)
                print("Bạn đã chiến thắng! Trò chơi kết thúc.")
                break

            else:
                current_player  = "human"

    play_again = input("Bạn có muốn chơi lại không? (y/n): ")
    if play_again == "y":
        continue
    else:
        break
```
