```python
import turtle
t = turtle.Turtle()

def draw_house():
    t.pencolor('blue')
    
    room_width = int(input("Nhập bề rộng căn phòng: "))
    room_height = int(input("Nhập chiều cao căn phòng: "))
    
    window_width = int(input("Nhập bề rộng cửa sổ: "))
    window_height = int(input("Nhập chiều cao cửa sổ: "))
    
    front = int(input("Nhập chiều dài cạnh mặt trước căn nhà: "))
    
    door_width = int(input("Nhập bề rộng cánh cửa: "))
    door_height = int(input("Nhập chiều cao cánh cửa: "))
    
    def draw_room(room_width, room_height):
        vi_tri_truoc_khi_ve = t.position()
        
        for i in range(2):
            t.fillcolor('red')
            t.begin_fill()
            t.forward(room_width)
            t.left(90)
            t.forward(room_height)
            t.left(90)
            t.end_fill()
            
        def ve_vi_tri_ban_dau():
            t.penup()
            t.goto(vi_tri_truoc_khi_ve)
            t.pendown()

        def draw_window(window_width, window_height):
            for i in range(2):
                t.fillcolor('white')
                t.begin_fill()
                t.forward(window_width)
                t.left(90)
                t.forward(window_height)
                t.left(90)
                t.end_fill()
            ve_vi_tri_ban_dau()
            
        t.penup()
        t.forward(room_width / 2 - window_width / 2)
        t.left(90)
        t.forward(room_height / 2 - window_height / 2)
        t.right(90)
        t.pendown()

        draw_window(window_width, window_height)
    
    draw_room(room_width, room_height)
    t.forward(room_width)
    
    draw_room(room_width, room_height)
    t.backward(room_width)
    t.right(90)
    t.forward(room_height)
    t.left(90)
    
    draw_room(room_width, room_height)
    t.forward(room_width)
    
    draw_room(room_width, room_height)
    t.forward(room_width)
    
    def ve_mat_truoc_can_nha():
        t.fillcolor('grey')
        t.begin_fill()
        for i in range(4):
            t.forward(front)
            t.left(90)
        t.end_fill()
        
        def ve_2_canh_cua():
            t.fillcolor('green')
            t.begin_fill()
            for i in range(2):
                for i in range(2):
                    t.forward(door_width)
                    t.left(90)
                    t.forward(door_height)
                    t.left(90)
                t.penup()
                t.forward(door_width)
                t.pendown()
            t.end_fill()
            
        t.forward(front / 2 - door_width)
        ve_2_canh_cua()
        
    ve_mat_truoc_can_nha()
    
draw_house()
```
