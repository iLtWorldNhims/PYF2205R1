```python
from turtle import *
from math import *

speed(0)

# Đưa bút đến vị trí để bắt đầu vẽ (tùy theo kích thước và hình dáng màn hình hiển thị hiện tại)
pu()
goto(-100, -100)
pd()

# Tạo các biến sẽ sử dụng để thuận tiện hơn khi vẽ các đoạn vẽ lặp lại. Và hơn hết là để các đối tượng sau khi vẽ sẽ được vẽ với tỉ lệ y chang nhau khi muốn vẽ ở các độ dài khác nhau mà tổng thể vẫn cân đối hài hoà
a = 200 # Đây là độ rộng của đáy căn nhà
b = a / 3 # Đây là độ rộng của tán lá cây lớn nhất của cây bé hơn (bên trái)

# Tạo hàm màu bút vẽ
def mau(x):
	pencolor(x)

# Tạo hàm tô màu
def tm(y):
	fillcolor(y)
	
mau("blue") # Đây sẽ là màu bút vẽ chủ đạo của bức tranh

# Vẽ mặt trước căn nhà
for i in range(2):
	fd(a)
	lt(90)
	fd(a/2.5)
	lt(90)

# Đưa bút đến vị trí sẽ vẽ cánh cửa
fd(a/5*2)
lt(90)

# Vẽ cánh cửa
for i in range(2):
	fd(a/3)
	rt(90)
	fd(a/5)
	rt(90)

# Đưa bút đến vị trí vẽ mái nhà
lt(90)
fd(a/5*2)
rt(90)
fd(a/5*2)

# Vẽ mái nhà
d = (a*(2**0.5)/3)
rt(45)
fd(d)
rt(90)
fd(d)

# Đưa bút đến vị trí vẽ ống khói
bk(d/5*3)
lt(135)

# Vẽ ống khói
fd(d/5)
rt(90)
fd(d/5)
rt(90)
fd(d/2.5)
bk(d/2.5)

# Đưa bút đến vị trí vẽ khói dài
pu()
bk(d/5)
pd()

# Vẽ khói dài
bk(d/2.5)

# Đưa bút đến vị trí vẽ khói ngắn
pu()
rt(90)
fd(d/5)
lt(90)
fd(d/10)
pd()

# Vẽ khói ngắn
fd(d/5)

# Đưa bút đến vị trí vẽ cái cây nhỏ (bên trái)
pu()
fd(a*1.5)
rt(90)
fd(a*1.3)
rt(90)
pd()

# Tạo hàm vẽ cây
def cay(b):
	mau("blue")
	
	# Vẽ thân cây
	begin_fill()
	tm('grey')
	rt(15)
	fd(b)
	rt(75)
	fd(b/4)
	rt(75)
	fd(b)
	rt(105)
	fd(2*(b/4+(b**2-(b*sin(radians(75)))**2)**0.5)-b/4)
	end_fill()
	
	# Đưa bút đến vị trí vẽ lá
	pu()
	rt(90)
	fd(b*sin(radians(75)))
	rt(90)
	fd(b/2.5)
	bk(b/2)
	lt(45)
	pd()
	
	# Đổi màu bút thành màu xanh lá 
	mau("chartreuse")
	
	# Vẽ 3 lá cây nhỏ dần và nhạt dần lên trên
	tm("dark green") # Lá dưới cùng
	begin_fill()
	c = b * (2**0.5) / 2
	fd(c)
	rt(90)
	fd(c)
	rt(135)
	fd(b)
	end_fill()
	
	pu()
	rt(90)
	fd(b/2)
	rt(90)
	fd(0.1*b)
	lt(45)
	pd()
	
	b *= 0.8
	tm("chartreuse") # Lá ở giữa
	begin_fill()
	c = b * (2**0.5) / 2
	fd(c)
	rt(90)
	fd(c)
	rt(135)
	fd(b)
	end_fill()
	
	rt(90)
	pu()
	fd(b/2)
	rt(90)
	fd(0.1*b)
	lt(45)
	pd()
	
	b *= 0.8
	tm("light green") # Lá trên cùng
	begin_fill()
	c = b * (2**0.5) / 2
	fd(c)
	rt(90)
	fd(c)
	rt(135)
	fd(b)
	end_fill()
		
# Vẽ cây nhỏ
cay(b)

# Đưa bút đến vị trí vẽ cây to
pu()
rt(90)
bk(a/2)
lt(90)
bk(a*2.3)
rt(90)
pd()

# Vẽ cây to
cay(b*1.5)

done()
```
