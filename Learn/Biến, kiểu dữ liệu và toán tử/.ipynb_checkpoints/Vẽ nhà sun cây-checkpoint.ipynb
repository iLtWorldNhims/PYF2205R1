```python
from turtle import *
bgcolor('turquoise')
speed(1)
pu()
goto(-300, -300)
pd()
a = 180
c = a * 2/3
# Ve ngoi nha
# tao ham to mau
def tm(x):
	fillcolor(x)
	
begin_fill()
for i in range(4): # ve mat truoc
	tm('blue')
	fd(a)
	lt(90)
end_fill()

fd(a/3)
lt(90)

begin_fill()
for i in range(2): # ve canh cua
	tm('green')
	fd(a/2)
	rt(90)
	fd(a/3)
	rt(90)
end_fill()

lt(90)
fd(a/3)
rt(90)
fd(a)

# ve mai nha
begin_fill()
tm('pink')
b = (a*(2**0.5)/2)
rt(45)
fd(b)
rt(90)
fd(b)
end_fill()

# ve mai nha ben hong
begin_fill()
tm('orange')
for i in range(4):
	lt(90)
	fd(b)
end_fill()

# ve hong nha
begin_fill()
tm('yellow')
lt(90)
fd(b)
rt(135)
fd(a)
rt(45)
fd(b)
end_fill()

bk(b/3)
rt(135)
pu()
fd(a/2)
pd()

# ve cua so
begin_fill()
tm('brown')
for i in range(2):
	fd(a/4)
	rt(45)
	fd(b/3)
	rt(135)
end_fill()

pu()
bk(a/2)
rt(90)
fd(a*1.7)
pd()

# Ve cai cay
# Ve than cay
begin_fill()
tm('brown')
for i in range(2):
	fd(c/5)
	lt(90)
	fd(c/2)
	lt(90)
end_fill()

lt(90)
fd(c/2)
rt(90)
bk(c/2.5)

# Ve la cay - 3 la - nho dan len tren
tm('green')
for i in range(3):
	begin_fill()
	d = c * (2**0.5) / 2
	lt(45)
	fd(d)
	rt(90)
	fd(d)
	rt(135)
	fd(c)
	rt(90)
	end_fill()
	pu()
	fd(c/2)
	rt(90)
	fd(c*0.05)
	pd()
	c *= 0.9

pu()
lt(90)
fd(a)
pd()

e = a/5
# Ve mat troi
begin_fill()
tm('yellow')
circle(e)
end_fill()
for i in range(4):
	rt(90)
	fd(e)
	bk(e)
	lt(90)
	circle(e, 45)
	rt(90)
	fd(e/2)
	bk(e/2)
	lt(90)
	circle(e, 45)

done()
```
